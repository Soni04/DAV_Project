```python
# Define a function to convert numeric indicators to diagnoses
def convert_to_diagnosis(row):
    diagnoses = []
    inconclusive = True
    for col, val in row.items():
        if val == 1:
            diagnoses.append(col)
            inconclusive = False
    if inconclusive:
        diagnoses.append('Inconclusive')
    return diagnoses

# Apply the function row-wise to convert the DataFrame to a list of diagnoses
diagnoses_list = results.apply(convert_to_diagnosis, axis=1)
diagnoses_df = pd.DataFrame(diagnoses_list.tolist(), columns=['Diagnosis'])

diagnoses_df
```
